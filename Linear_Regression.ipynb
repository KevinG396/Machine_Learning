# Linear Regression

### Dataset

```python
import numpy as np
from matplotlib import pyplot as plt

np.random.seed(42)
# Generate x values evenly spaced between 1 and 10
x = np.linspace(1, 10, 50)
# Generate random noise with a normal distribution (mean=0, std=1)
noise = np.random.normal(0, 1, 50)

y = 5*x + 15+noise
plt.scatter(x, y, label='Data Points')
plt.show()
```

### Least Squares Line

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
beta_1 = np.sum((x - x_mean) * (y - y_mean)) / np.sum((x - x_mean) ** 2)
beta_0 = y_mean - beta_1 * x_mean
y_hat = beta_0 + beta_1 * x
residuals = y - y_hat
ssr = np.sum(residuals ** 2)

plt.scatter(x, y, label='Data Points')
plt.plot(x, y_hat, color='red', label='Least Squares Line')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
print(beta_0,beta_1)
print(f"Sum of Squares of Residuals: {ssr}")
```
